```python
import pandas as pd
import seaborn as sns
```

```python
dataset_names=sns.get_dataset_names()
dataset_names
```

```python
tips_data=sns.load_dataset("tips")
tips_data
```

```python
sns.lineplot(x='day',y='total_bill',data=tips_data,hue="sex")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips_data,hue="sex")
```

```python
sns.barplot(x="day",y="tip",data=tips_data,hue="sex",palette="dark:orange")
```

```python
#barplot
# violinplot
# catplot
# histplot
# lineplot
```

```python
sns.catplot(x="day",y="tip",data=tips_data,order=["Sun","Sat"])
```

```python
sns.histplot(x="day",y="tip",data=tips_data)
```
