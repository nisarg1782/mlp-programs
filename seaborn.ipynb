```python
tips_data=sns.load_dataset('tips')
tips_data
```

```python
sns.lineplot(x='day',y='total_bill',data=tips_data,hue='sex')
```

```python
sns.boxplot(x='total_bill',y='sex',data=tips_data,color='red',hue='day')
```

```python
sns.barplot(x='total_bill',y='day',data=tips_data,hue='sex',palette='dark:blue',order=['Sun','Sat','Fri','Thr'])
```

```python
sns.histplot(y='day',x='total_bill',data=tips_data)
```
