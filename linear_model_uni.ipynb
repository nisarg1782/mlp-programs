```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("housing.csv")
data.head(3)
```

```python
x=data.drop("Price",axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,data["Price"],test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg1=LinearRegression()
```

```python
lg1.fit(x_train,y_train)
```

```python
y_predicts=lg1.predict(x_test)
```

```python
y_predicts.ndim
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_predicts,y_test)
```

```python
lg1.accuracy_score(x_test,y_test)
```
